```python
import time

# Initialize rocket state
rocket = {
    "stage": "Stage 1",
    "fuel": 100,
    "altitude": 0,
    "speed": 0
}

# Define parameters for each stage
stages = {
    #"Pre-Launch": {"fuel_depletion_rate": 0.0, "altitude_increase_rate": 0.0, "speed_increase_rate": 0.0},
    "Stage 1": {"fuel_depletion_rate": 1.0, "altitude_increase_rate": 10.0, "speed_increase_rate": 1000.0},
    "Stage 2": {"fuel_depletion_rate": 0.5, "altitude_increase_rate": 15.0, "speed_increase_rate": 1500.0},
}

# Function to update rocket parameters
def update_rocket_parameters(rocket):
    if rocket["stage"] in stages:
        stage_data = stages[rocket["stage"]]
        rocket["fuel"] -= stage_data["fuel_depletion_rate"]
        rocket["altitude"] += stage_data["altitude_increase_rate"]
        rocket["speed"] += stage_data["speed_increase_rate"]

# Function to display rocket parameters
def display_parameters(rocket):
    print(f"Stage: {rocket['stage']}, Fuel: {rocket['fuel']}%, Altitude: {rocket['altitude']} km, Speed: {rocket['speed']} km/h")

# Function to check for stage separation
def check_stage_separation(rocket):
    if rocket["stage"] == "Stage 1" and rocket["altitude"] >= 30:
        print("Stage 1 complete. Separating stage. Entering Stage 2.")
        rocket["stage"] = "Stage 2"

# Function to check if orbit is achieved
def check_orbit_achieved(rocket):
    if rocket["altitude"] >= 200 and rocket["speed"] >= 2000:
        print("Orbit achieved! Mission Successful.")
        rocket["stage"] = "Orbit"

# Function to handle mission result
def handle_mission_result(rocket):
    if rocket["stage"] == "Orbit":
        print("Mission Successful.")
    else:
        print("Mission Failed due to insufficient fuel.")

# Main simulation loop
while True:
    user_input = input("Enter a command: ")

    if user_input == "start checks":
        # Check all systems and report if 'Go' for launch
        print("All systems are 'Go' for launch.")

    elif user_input == "launch":
        # Start the mission and enter the simulation loop
        while rocket["stage"] != "Orbit" and rocket["fuel"] > 0:
            # Update rocket parameters
            update_rocket_parameters(rocket)
            # Display updated parameters
            display_parameters(rocket)
            # Check for stage separation or orbit
            check_stage_separation(rocket)
            check_orbit_achieved(rocket)
        # Handle mission success or failure
        handle_mission_result(rocket)

    elif user_input.startswith("fast forward "):
        try:
            seconds = int(user_input.split()[2])
            for _ in range(seconds):
                update_rocket_parameters(rocket)
                display_parameters(rocket)
                check_stage_separation(rocket)
                check_orbit_achieved(rocket)
                if rocket["stage"] == "Orbit" or rocket["fuel"] <= 0:
                    handle_mission_result(rocket)
                    break
        except ValueError:
            print("Invalid input for fast forward.")

    elif user_input == "exit":
        print("Exiting simulation.")
        break
```
